````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from io import StringIO
````

````python
sns.set_style('darkgrid')
````

````python
# Function to load and process the uploaded file
def load_data(file_path):
    try:
        # Try reading CSV file
        df = pd.read_csv(file_path)
        print("File loaded successfully!")
        print("\nDataset Info:")
        print(df.info())
        return df
    except Exception as e:
        print(f"Error loading file: {e}")
        print("Ensure the file is uploaded to JupyterLab's working directory or provide the correct path.")
        return None

````

````python
# Function to create histogram for numerical columns
def plot_histogram(df, column, bins=30):
    plt.figure(figsize=(10, 6))
    plt.hist(df[column].dropna(), bins=bins, edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(True, alpha=0.3)
    plt.show()
````

````python
# Function to create box plot for numerical columns
def plot_boxplot(df, column):
    plt.figure(figsize=(10, 6))
    sns.boxplot(y=df[column].dropna())
    plt.title(f'Box Plot of {column}')
    plt.ylabel(column)
    plt.grid(True, alpha=0.3)
    plt.show()
````

````python
# Function to create scatter plot for two numerical columns
def plot_scatter(df, x_col, y_col):
    plt.figure(figsize=(10, 6))
    plt.scatter(df[x_col], df[y_col], alpha=0.5)
    plt.title(f'Scatter Plot: {x_col} vs {y_col}')
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.grid(True, alpha=0.3)
    plt.show()
````

````python
# Main analysis function
def analyze_data(file_path):
    # Load the data
    df = load_data(file_path)
    
    if df is None:
        return
    
    # Display basic statistics
    print("\nBasic Statistics:")
    print(df.describe())
    
    # Get numerical columns
    numerical_cols = df.select_dtypes(include=[np.number]).columns
    
    # Create visualizations for numerical columns
    for col in numerical_cols:
        print(f"\nVisualizations for {col}:")
        
        # Histogram
        plot_histogram(df, col)
        
        # Box plot
        plot_boxplot(df, col)
        
        # Scatter plot (if there are at least two numerical columns)
        if len(numerical_cols) > 1 and col != numerical_cols[-1]:
            plot_scatter(df, col, numerical_cols[list(numerical_cols).index(col)+1])
````

````python
# Example usage
# Replace 'your_file.csv' with the actual path to your uploaded CSV file
file_path = 'processed_data.csv'  # Update this to the correct file name (e.g., 'data.csv')
analyze_data(file_path)

# Note: To use this notebook in JupyterLab:
# 1. Upload your CSV file to JupyterLab:
#    - In the JupyterLab web interface, click the upload button (upward arrow) in the file browser.
#    - Select your CSV file and upload it to the working directory.
# 2. Update 'your_file.csv' above to match the uploaded file name (e.g., 'data.csv').
# 3. If the file is in a subdirectory, include the path (e.g., 'data/my_file.csv').
# 4. The file should contain numerical columns for statistical analysis.
# 5. If you need to install dependencies, run the following in a notebook cell:
#    ```python
#    !pip install pandas==2.2.0 matplotlib==3.8.2 seaborn==0.13.2 numpy==1.26.0
#    ```
# 6. For non-CSV files (e.g., Excel), modify the load_data function, e.g.:
#    ```python
#    df = pd.read_excel(file_path)  # Requires openpyxl or xlrd
#    ```
````
